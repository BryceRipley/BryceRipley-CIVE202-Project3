```python
# CHUNK 1: Import Packages, Set The Plot Style, and Load the Data in.

# Step 1: Import the important packages used.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Set a clean visual style for all plots.

sns.set_style('whitegrid')
sns.set_palette('deep')

# Step 3: Load both the NHTS and NGSIM datasets and display the first few rows of data to ensure data is loaded properly.

nhts_data = pd.read_csv('NHTS.csv')

ngsim_data = pd.read_csv('NGSIM.csv')

print("NHTS Data Preview:")
display(nhts_data.head())

print("\nNGSIM Data Preview:")
display(ngsim_data.head())

# Step 4: Print dataset information for quick inspection

print("\nNHTS Dataset Info:")
print(nhts_data.info())

print("\nNGSIM Dataset Info:")
print(ngsim_data.info())

# Step 5: Summary statistics for numeric columns
print("\nNHTS Summary Statistics:")
display(nhts_data.describe())

print("\nNGSIM Summary Statistics:")
display(ngsim_data.describe())
```

```python
# CHUNK 2: Clean and Prepare NHTS Data

# Step 1: Remove outliers in vehicle_age (> 30 years)

nhts_data = nhts_data[nhts_data['vehicle_age'] <= 30]


# Step 2: Drop rows with missing values in key variables
nhts_data = nhts_data.dropna(subset=['household_income','vehicles_per_household','vehicle_age','urban'])

# Step 3: Preview cleaned data

print("NHTS Data After Cleaning:")
display(nhts_data.head())

print("\nRemaining Missing Values:")
print(nhts_data.isnull().sum())
```

```python
# CHUNK 3: NHTS Visualizations

# BAR CHART — Vehicles per Household by Income Category

# Step 1: Compute the mean vehicles per household for each income group

income_means = nhts_data.groupby('household_income')['vehicles_per_household'].mean()

# Step 2: Create the bar chart 
plt.figure(figsize = (12,6))

income_means.plot(kind = 'bar',color = sns.color_palette('Set2', len(income_means)),edgecolor = 'black')

# Step 3: Customize labels and title
plt.xlabel('Household Income Category', fontsize = 12)
plt.ylabel('Average Vehicles per Household', fontsize = 12)
plt.title('Vehicles per Household by Income Category', fontsize = 18)
plt.xticks(rotation = 45, fontsize = 10)

# Step 4: Show the plot
plt.tight_layout()
plt.show()

# HISTOGRAM of Vehicle Age Distribution

# Step 1: Create a histogram to visualize the distribution of vehicle ages in the NHTS dataset

plt.figure(figsize=(10,6))

sns.histplot(nhts_data['vehicle_age'], bins=20, kde=True, color='skyblue', edgecolor='black')

# Add axis labels and title.
plt.xlabel('Vehicle Age (years)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Vehicle Age', fontsize=18)

# Step 3: Display the plot.
plt.tight_layout()
plt.show()


# BOXPLOT of Vehicles per Household by Urban Status

#Step 1: Cretae a boxplot to compare the distribution of vehicles per 
# household between urban and non-urban households

plt.figure(figsize=(10,6))

sns.boxplot(data=nhts_data, x='urban', y='vehicles_per_household', palette='Set3')

# Step 2: Add axis labels and title.
plt.xlabel('Urban Status (0 = Non-Urban, 1 = Urban)', fontsize=12)
plt.ylabel('Vehicles per Household', fontsize=12)
plt.title('Vehicles per Household by Urban vs Non-Urban Households', fontsize=18)

# Step 3: Display the plot.
plt.tight_layout()
plt.show()
```

```python
# CHUNK 4: Clean and Prepare NGSIM Data

# Step 1: Reset time so that it starts at 0 seconds
ngsim_data['Time'] = ngsim_data['Time'] - ngsim_data['Time'].min()

# Step 2: Compute the gap distance between leader and follower
# gap = leader_position - follower_position
ngsim_data['gap_distance'] = ngsim_data['leader_position(m)'] - ngsim_data['follower_position(m)']

# Step 3: Drop rows with missing values in key variables
ngsim_data = ngsim_data.dropna(subset=['Time','leader_speed(m/s)','follower_speed(m/s)','leader_position(m)','follower_position(m)',
                                       'leader_acc(m/s^2)','follower_acc(m/s^2)','gap_distance'])

# Step 4: Select a trajectory number for time-series plots
trajectory_number = 4
ngsim_subset = ngsim_data[ngsim_data['trajectory_number'] == trajectory_number].reset_index(drop=True)

# Step 5: Preview the cleaned subset
print("NGSIM Subset Preview (Trajectory 4):")
display(ngsim_subset.head())

print("\nRemaining Missing Values in NGSIM:")
print(ngsim_data.isnull().sum())
```

```python
# CHUNK 5: NGSIM Time-Series Plots

# 1. Follower Speed vs Time
plt.figure(figsize = (12,6))

# Step 1: Plot follower speed over time
plt.plot(ngsim_subset['Time'], ngsim_subset['follower_speed(m/s)'], color = 'steelblue', linewidth = 2, label = 'Follower Speed')
         
# Step 2: Customize Labels and Title + Legend
plt.xlabel('Time (seconds)', fontsize = 12)
plt.ylabel('Follower Speed (m/s)', fontsize = 12)
plt.title('Follower Speed vs Time', fontsize = 18)
plt.legend()

# Step 4: Show the plot
plt.tight_layout()
plt.show()

# 2. Follower Position vs Time
plt.figure(figsize = (12,6))

# Step 1: Plot follower position over time
plt.plot(ngsim_subset['Time'], ngsim_subset['follower_position(m)'], color = 'darkgreen', linewidth = 2, label = 'Follower Position')

# Step 2: Customize Labels and Title + Legend
plt.xlabel('Time (seconds)', fontsize = 12)
plt.ylabel('Follower Position (m)', fontsize = 12)
plt.title('Follower Position vs Time', fontsize = 18)
plt.legend()

# Step 4: Show the plot
plt.tight_layout()
plt.show()




```
